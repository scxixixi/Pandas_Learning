### 一. 单级索引

1. loc方法、iloc方法、[]操作符
最常用的索引方法可能就是这三类，其中iloc表示位置索引，loc表示标签索引，[]也具有很大的便利性，各有特点

### (1)loc 左闭右闭

```python
# 单行索引
df.loc[1101] # return: Series
df.loc[[1101]] # return: DataFrame
```

```python
# loc 是按照标签索引 (key)
df.loc[[1101,1102]]
```

```python
# loc中使用的切片区间为左右全闭 切片不支持末尾 -1 的使用
df.loc[1105::2].head()
```

### 3) [ ]操作符

#### 3.1) Series的 [ ] 操作符 

```python
# 单元素索引
type(df['Gender']) #Series
s = pd.Series(df['Math'])#,index = df.index,name = 'Math')
s[1101]
```

```python
# 切片索引
s[0:3] #使用的是绝对位置的整数切片，与元素无关，这里容易混淆
```

```python
#函数式索引 我吐了（T-T）
s[lambda x: x.index[16::-6]] # 对比切片索引（绝对位置），函数式（位置）
#注意使用lambda函数时，直接切片(如：s[lambda x: 16::-6])就报错，此时使用的不是绝对位置切片，而是元素切片，非常易错
```

```python
# 布尔索引
s>80 # Series
s[s>80] # Bool Series Ok loc/[]
s[pd.Series(s>80).values] # Bool List Ok iloc/loc/[]
```

```python
s_int = pd.Series([1,2,3,4],index=[1,3,5,6])
s_float = pd.Series([1,2,3,4],index=[1.,3.,5.,6.])
s_float
```

## 2. 布尔索引

### (a) 布尔符号：'&','|','~'：分别代表和and，或or，取反not

```python
df[(df['Gender']=='F')&(df['Address']=='street_2')].head()
```

```python
# loc和[]中相应位置都能使用布尔列表选择
df.loc[df['Math']>60,df.columns=='Physics'].head()
```

```python
#思考：为什么df.loc[df['Math']>60,(df[:8]['Address']=='street_6').values].head()得到和上述结果一样？values能去掉吗？
df.loc[df['Math']>60,(df[:8]['Address']=='street_6').values].head()
```

```python
(df[:8]['Address']=='street_6').values
```

```python
# 答：因为以上返回的是布尔数组，放在第二个位置，标明对列的选择。不可把values去掉，去掉后和列的大小不一致
```

```python
df[df['Address'].isin(['street_1','street_4'])&df['Physics'].isin(['A','A+'])]
```

```python
#上面也可以用字典方式写：
df[df[['Address','Physics']].isin({'Address':['street_1','street_4'],'Physics':['A','A+']}).all(1)]
#all与&的思路是类似的，其中的1代表按照跨列方向判断是否全为True
```

```python
df1 = df.loc[df[['Address','Physics']].isin({'Address':['street_1','street_4'],'Physics':['A','A+']}).all(1)]
df1
```

```python
df1[['Class','Address','Physics']].isin({'Address':['street_1','street_4'],'Physics':['A','A+']}).all(0)
```

1是跨列去寻找，也就是按照符合条件的列去寻找对应的行。 0是跨行寻找，也就是按照符合条件的行去寻找对应的列。而在这个例子中，Class这一列无法满足条件，而Address可以满足条件

## 4. 区间索引

### a）利用interval_range方法

```python
pd.interval_range(start=0,end=5)
#closed参数可选'left''right''both''neither'，默认左开右闭
```

```python
pd.interval_range(start=0,periods=8,freq=5,closed='both')
#periods参数控制区间个数，freq控制步长
```

### b）利用cut将数值列转为：区间为元素的分类变量，例如统计数学成绩的区间情况：

```python
math_interval = pd.cut(df['Math'],bins=[0,40,60,80,100])
#注意，如果没有类型转换，此时并不是区间类型，而是category类型
math_interval.head()
```

### c）区间索引的选取

```python
df_i = df.join(math_interval,rsuffix='_interval')[['Math','Math_interval']]\
            .reset_index().set_index('Math_interval')
df_i.head()
```

```python
df.join(math_interval,rsuffix='_interval')[['Math','Math_interval']].head()
```

```python
df_i.loc[65].head()
#包含该值就会被选中
```

```python
df_i.loc[[65,85]].head()
```

#### 如果想要选取某个区间，先要把分类变量转为区间变量，再使用overlaps方法：

```python
#df_i.loc[pd.Interval(70,75)].head() 报错
df_i[df_i.index.astype('interval').overlaps(pd.Interval(70, 85))].head()
#只要索引与(70,85]这个区间有交集就会被选中，注意pd.Interval默认构造区间都是左开右闭，可选closed参数right,left,both,neither
```
